```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

# READING IMAGE

```python
img = cv2.imread("C:\\Users\\aedpu\\Gurubalan Final year\\C.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img 
```

# RDB COVERSION

```python
imrgb =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

# RESHAPE

```python
vec = imrgb.reshape((-1,3))
vec.shape
```

# FLOAT CONVERSION

```python
vec = np.float32(vec)
```

# iteration termination criiteria

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,100,1 ) 
```

# CLUSTER CONDITIONS

```python
k = 4
attempts = 100
```

# k- MEANS CLUSTERING

```python
ret,label,center = cv2.kmeans(vec,k,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
```

# UNIT CONVERSION

```python
center = np.uint8(center)
```

# Flattening and Reshape

```python
res = center[label.flatten()]
res_img = res.reshape((imrgb.shape))
```

# original image

```python
cv2.imshow('original',img)
cv2.imshow('segmented' ,res_img) 
    
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('segmented.jpg',res_img)
```
