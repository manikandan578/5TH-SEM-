sobel operator

```python
img = cv2.imread("C:\\Users\\aedpu\\MANIKANDAN\\table.png")
```

```python
img
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
```

```python
sobelX = cv2.Sobel(img,cv2.CV_64F,1,0)
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1)
```

```python
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))
```

```python
combind = cv2.bitwise_or(sobelX,sobelY)
```

```python
plt.imshow(sobelY)
```

```python
plt.imshow(sobelX)
```

laplacian operactor

```python
lab = cv2.Laplacian(img, cv2.CV_64F,ksize=3)
lab = np.uint8(np.absolute(lab))
```

Display Results

```python
titles = ['image ','sobelX','sobelY','combind','Laplacian']
image = [img,sobelX,sobelY,combind,lab]

for i in range(5):
    plt.subplot(2,3,i+1),plt.imshow(image[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
