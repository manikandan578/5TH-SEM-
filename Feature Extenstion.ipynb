```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
from skimage.feature import hog
from skimage import data ,exposure
```

# READING IMG

```python
img_1 = cv2.imread("C:\\Users\\aedpu\Downloads\\ggg.jpg")
```

```python
img_1
```

```python
img = cv2.resize(img_1,(512,512))
```

# HOG

```python
cv2.imshow('original',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
featdis , hog_image = hog(img , pixels_per_cell = (8, 8 ),
                         cells_per_block = (1,1 ) , visualize = True ,channel_axis= 2)
```

```python
print(featdis)
```

```python
cv2.imshow('hog',hog_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Rescaling

```python
rescale = exposure.rescale_intensity(hog_image, in_range=(0,1))
```

```python
rescale
```

# plot

```python
fingure , (a1,a2) = plt.subplots(1,2,figsize = (12,6) , sharex = True , sharey = True )


a1.axis('off')
a1.imshow(img)
a1.set_title('input image')

a2.axis('off')
a2.imshow(rescale)
a2.set_title("Histogram of Oriented Gradients")
plt.show()
```
