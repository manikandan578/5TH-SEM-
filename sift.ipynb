```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread("light1.png")
image.shape
```

```python
image = cv2.resize(image,(512,512))
image.shape
```

```python
cv2.imshow("img",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#color conversion
rgb= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
                  
```

```python
#scale  invariance
sc_iv = cv2.pyrDown(rgb)
sc_iv2 = cv2.pyrDown(sc_iv)
num_rows,num_cols = sc_iv2.shape[:2]
print(num_rows)
print(num_cols)
```

```python
cv2.imshow("img",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow("img",sc_iv)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow("img",sc_iv2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#rotational invariance
rot_iv = cv2.getRotationMatrix2D((num_rows/2,num_cols/2),45,1)
```

```python
aff = cv2.warpAffine(sc_iv2,rot_iv,(num_rows,num_cols ))
```

```python
aff
```

Color Convertion

```python
rot_gray = cv2.cvtColor(aff,cv2.COLOR_RGB2GRAY)
```

ROTIATION DISPLAY

```python
fx , plots = plt.subplots(1,2,figsize=(20,10))

plots[0].set_title("training image 1")
plots[0].imshow(rgb)

plots[1].set_title("training image 2")
plots[1].imshow(aff)
plt.show()
```

color change

```python
gray =  cv2.cvtColor(aff,cv2.COLOR_RGB2GRAY)
```

sift

```python
gg = cv2.xfeatures2d.SIFT_create()
```

Detect and Compute

```python
train_gg,train_des=gg.detectAndCompute(sc_iv,None)
test_gg,test_dos = gg.detectAndCompute(rot_gray,None)
```

```python
print(test_gg)
```

```python
gg_wo_size = np.copy(rgb)
gg_w_size = np.copy(rgb)
```

Drawing keypoints

```python
cv2.drawKeypoints(sc_iv, train_gg, gg_w_size,color = (0,255,0))
cv2.drawKeypoints(sc_iv, train_gg, gg_w_size, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

KEY POINT DISPLAY

```python
fx, plots = plt.subplots(1,2,figsize=(20,10))

plots[0].set_title("training keypoints")
plots[0].imshow(gg_w_size, cmap='gray')

plots[1].set_title("train keypoints without Size")
plots[1].imshow(gg_wo_size, cmap='gray')
plt.show()
```

Brute Force Matcher Object

```python
bfm =cv2.BFMatcher(cv2.NORM_L1,crossCheck=False)
```

MATCHING

```python
match = bfm.match(train_des, test_dos)
```

Sorting

```python
match = sorted(match, key = lambda x : x.distance)
```

Drawing Matches

```python
result = cv2.drawMatches(sc_iv,train_gg,rot_gray , test_gg,match,rot_gray,flags=2)
```

Matching Display

```python
plt.rcParams['figure.figsize'] = [5.5,5.5]
plt.title('best macthing points')
plt.imshow(result)
plt.show()
```
